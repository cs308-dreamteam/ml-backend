```python
import random
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from faker import Faker
import uuid
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn as sns
```

```python
fake = Faker()
artists = ['Artist A', 'Artist B', 'Artist C', 'Artist D', 'Artist E', 'Artist F', 'Artist G', 'Artist H', 'Artist I', 'Artist J']

music_genres = ['Pop', 'Rock', 'Hip Hop', 'Country', 'Electronic', 'R&B', 'Jazz', 'Classical', 'Indie', 'Folk']


user_data = {
    'song_id': [],
    'song_name': [],
    'artist': [],
    'album': [],
    'release_date': [],
    'added_date': [],
    'popularity': [],
    'liked': [],
    'rating': [],
    'genre': []
}

for artist in artists:
    for album_num in range(1, 11):
        album_name = f'{artist} Album {album_num}'
        album_release_date = datetime(1970, 1, 1) + timedelta(days=np.random.randint(0, 365 * 50))

        for song_num in range(1, 6):
            song_id = uuid.uuid4()
            song_name = fake.word()
            release_date = album_release_date
            added_date = datetime(2023, 1, 1) - timedelta(days=np.random.randint(0, 365*3))
            popularity = np.random.randint(1, 101)
            liked = np.random.choice([True, False])
            rating = np.random.uniform(0.0, 5.0,)
            genre = np.random.choice(music_genres)

            user_data['song_id'].append(song_id)
            user_data['song_name'].append(song_name)
            user_data['artist'].append(artist)
            user_data['album'].append(album_name)
            user_data['release_date'].append(release_date)
            user_data['added_date'].append(added_date)
            user_data['popularity'].append(popularity)
            user_data['liked'].append(liked)
            user_data['rating'].append(round(rating, 1))
            user_data['genre'].append(genre)


df = pd.DataFrame(user_data)

df = df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
def get_top_albums_by_date(user_data, start_date, end_date):
    user_data['release_date'] = pd.to_datetime(user_data['release_date'])

    # Belirli zaman aralığındaki albümleri filtrele
    filtered_df = user_data[(user_data['release_date'] >= start_date) &
                             (user_data['release_date'] <= end_date)]

    # Her albümün ortalama rating'ini ve çıkış tarihi
    album_info = filtered_df.groupby('album').agg({'rating': 'mean', 'release_date': 'max'}).reset_index()

    top_albums = album_info.nlargest(10, 'rating')

    return top_albums[['album', 'rating', 'release_date']]

# Örnek olarak bir zaman aralığı belirle
start_date = datetime(1950, 1, 1)
end_date = datetime(2023, 1, 1)

# Belirli zaman aralığındaki en çok sevilen 5 albümü bul
top_albums = get_top_albums_by_date(df, start_date, end_date)

print(top_albums)
```

```python
def get_top_artists_by_rating(user_data):
    # Her sanatçının ortalama rating'ini hesapla
    artist_info = user_data.groupby('artist')['rating'].mean().reset_index()

    top_artists = artist_info.nlargest(10, 'rating')

    return top_artists[['artist', 'rating']]

top_artists = get_top_artists_by_rating(df)

print(top_artists)
```

```python
def get_favorite_songs_in_date_range(user_data, start_date, end_date):

    filtered_df = user_data[(user_data['added_date'] >= start_date) &
                             (user_data['added_date'] <= end_date)]

    top_songs = filtered_df.nlargest(10, 'rating')

    return top_songs[['song_name', 'rating', 'release_date', 'added_date']]

start_date = datetime(2000, 1, 1)
end_date = datetime(2022, 12, 31)

favorite_songs_in_date_range = get_favorite_songs_in_date_range(df, start_date, end_date)

print(favorite_songs_in_date_range)
```

```python
def get_favorite_songs_in_date_range(user_data, start_date, end_date):
    # Şarkıları belirli bir zaman aralığına göre filtrele
    filtered_df = user_data[(user_data['added_date'] >= start_date) &
                             (user_data['added_date'] <= end_date)]

    # En çok beğenilen 10 şarkıyı seç
    top_songs = filtered_df.nlargest(10, 'rating')

    return top_songs[['song_id', 'song_name', 'rating', 'release_date', 'added_date']]

# Örnek olarak bir zaman aralığı belirle
start_date = datetime(2000, 1, 1)
end_date = datetime(2022, 12, 31)

# Belirli bir zaman aralığındaki en çok beğenilen 10 şarkıyı bul
favorite_songs_in_date_range = get_favorite_songs_in_date_range(df, start_date, end_date)

print(favorite_songs_in_date_range)
```

```python
def plot_user_song_genre_distribution(user_data):
    # Kullanıcının eklediği şarkıların genre'larına göre sayısını hesapla
    genre_counts = user_data['genre'].value_counts()

    # Renk paletini ayarla
    colors = sns.color_palette('pastel')[0:len(genre_counts)]

    # Pasta grafiği oluştur
    plt.figure(figsize=(10, 8))
    plt.pie(genre_counts, labels=genre_counts.index, autopct=lambda p: '{:.0f}\n({:.1f}%)'.format(p * sum(genre_counts) / 100, p),
            startangle=90, colors=colors, wedgeprops=dict(width=0.4))

    # Yazı tipi ve boyutunu ayarla
    plt.title('Kullanıcının Eklediği Şarkıların Türleri Dağılımı', fontsize=16, fontweight='bold')
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

    plt.show()

# Kullanıcının eklediği şarkıların genre dağılımını göster
plot_user_song_genre_distribution(df)
```

```python
def plot_genre_average_rating(user_data):
    # Her genre için ortalama rating
    genre_avg_ratings = user_data.groupby('genre')['rating'].mean().sort_values(ascending=False)

    #Renk paleti
    colors = sns.color_palette('pastel')[0:len(genre_avg_ratings)]

    #Graph
    plt.figure(figsize=(12, 6))
    sns.barplot(x=genre_avg_ratings.index, y=genre_avg_ratings.values, palette=colors, edgecolor=".6")

    #Grafik üzerindeki değerleri göster
    for i, value in enumerate(genre_avg_ratings.values):
        plt.text(i, value + 0.05, round(value, 2), ha='center', va='bottom', fontsize=10, color='black')

    # Yazı tipi ve boyutunu ayarla
    plt.title('Her Genre için Ortalama Rating', fontsize=16, fontweight='bold')
    plt.xlabel('Genre', fontsize=14)
    plt.ylabel('Ortalama Rating', fontsize=14)
    plt.xticks(rotation=45, ha='right', fontsize=12)
    plt.yticks(fontsize=12)

    plt.show()

plot_genre_average_rating(df)
```
